## Pandas Export
- to csv
- to excel
- to html
- to json
- to sql

```python
import pandas as pd
df = pd.DataFrame()
```

## to_csv

```python
df = pd.read_csv('deliveries.csv')
df.head()
```

```python
df.groupby('batsman')['batsman_runs'].sum()
```

```python
temp_df = df.groupby('batsman')['batsman_runs'].sum().reset_index()
```

```python
temp_df
```

```python
temp_df.to_csv('bastmna_runs.csv',index=False)
```

```python
df.pivot_table(index='batsman',columns='bowling_team',values='batsman_runs',aggfunc='sum').to_csv('bats')
```

## to_excel

```python
df
```

```python
temp_df = df.groupby('batsman')['batsman_runs'].sum().reset_index()
```

```python
temp_df.to_excel('batsman_runs.xlsx')
```

## to_html

```python
df
```

```python
df.query('batsman_runs ==6').pivot_table(index='over',columns='ball',values='batsman_runs',aggfunc='count')
```

```python
df.query('batsman_runs ==6').pivot_table(index='over',columns='ball',values='batsman_runs',aggfunc='count').to_html('sixes_heatmap.html')
```
